```python
import json
import pandas as pd
import glob


folder_path = "../sensor_data/"
```

```python
sensor_data_files = glob.glob(folder_path + "*")
```

```python
sensor_data_files
```

```python
import plotly.express as px


for fl in sensor_data_files:
    with open(fl, "r") as f:
        sensor_data = json.load(f)
        coordinates = []
        for d in sensor_data:
            if "gnss" in d and 'location' in d['gnss']:
                coords = d['gnss']["location"]['coordinates']
                coordinates.append({'lat': coords[0], 'long': coords[1]})
        df = pd.DataFrame(coordinates)
        fig = px.scatter_mapbox(df,lat='lat', lon='long')
        fig.update_layout(mapbox_style='open-street-map')
        fig.show()
```
